```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_nodes_from(['a', 'b', 'c', 'd'])
G.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('a','c')])
nx.draw(G, with_labels = True)
```

```python
list(G.neighbors('b'))
```

```python
for neighbor in G.neighbors('b'):
    print(neighbor)
```

```python
G.has_node('a')
```

```python
nx.is_tree(G)
```

```python
nx.is_connected(G)
```

```python
G.degree('a')
```

```python
D = nx.DiGraph()
D.add_edges_from([('1','2'), ('2','3'), ('1','4'), ('3','4')])
nx.draw(D, with_labels = True)
```
